## Visualization & Explanatory Analysis

### Q 1 : Apa saja Produk yang paling laris dan paling tidak laris?

```python
sum_order_items_df = all_data.groupby("product_category_name_english")["product_id"].count().reset_index()
sum_order_items_df = sum_order_items_df.rename(columns={"product_id": "products"})
sum_order_items_df = sum_order_items_df.sort_values(by="products", ascending=False)
sum_order_items_df = sum_order_items_df.head(10)

sum_order_items_df.head()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))  # Mengubah ukuran gambar

colors = ["#53ECEC", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]  # Mengubah warna

# Plot untuk 5 produk teratas
sns.barplot(x="products", y="product_category_name_english", data=sum_order_items_df.head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Top 5 Best Selling Products (Laku)", loc="center", fontsize=16)
ax[0].tick_params(axis='y', labelsize=12)

# Plot untuk 5 produk terbawah
sns.barplot(x="products", y="product_category_name_english", data=sum_order_items_df.sort_values(by="products", ascending=True).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Top 5 Least Selling Products(≠Laku)", loc="center", fontsize=16)
ax[1].tick_params(axis='y', labelsize=12)

# Menambahkan stroke hitam pada warna #53ECEC untuk ax[0] dan ax[1]
for bar in ax[0].patches:
    bar.set_edgecolor('black')
    bar.set_linewidth(1.5)

for bar in ax[1].patches:
    bar.set_edgecolor('black')
    bar.set_linewidth(1.5)

plt.suptitle("Produk Paling Laku dan Sedikit Terjual", fontsize=20)
plt.show()
```

Dari grafik di atas, dapat dilihat bahwa kategori produk yang paling banyak terjual adalah 'Bed & Bath Table', sementara yang paling sedikit terjual adalah 'Auto'.

### Q 2 : Bagaimana performa penjualan pada platform E-Commerce ?

```python
# Mengelompokkan data berdasarkan bulan dan menghitung jumlah unik pesanan setiap bulannya
monthly_df = all_data.resample(rule='M', on='order_approved_at').agg({
    "order_id": "nunique",
})

# Mengubah format indeks dari tanggal menjadi nama bulan
monthly_df.index = monthly_df.index.strftime('%B')

# Mengatur ulang indeks dan mengubah nama kolom
monthly_df = monthly_df.reset_index().rename(columns={"order_id": "order_count"})
monthly_df.head()
```

```python
monthly_df = monthly_df.sort_values('order_count').drop_duplicates('order_approved_at', keep='last')
```

```python
monthly_df.head()
```

```python
monthly_df.sort_values(by='order_count')
```

```python
month_mapping = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12
}

monthly_df["month_numeric"] = monthly_df["order_approved_at"].map(month_mapping)
monthly_df = monthly_df.sort_values("month_numeric")
monthly_df = monthly_df.drop("month_numeric", axis=1)
```

```python
plt.figure(figsize=(12, 6))  # Mengubah ukuran gambar

plt.plot(
    monthly_df["order_approved_at"],
    monthly_df["order_count"],
    marker='o',
    linewidth=2,
    color="#53ECEC",  # Mengubah warna garis plot
    linestyle='-',  # Menambahkan garis lurus
    mec='black',  # Menambahkan warna garis tepi hitam
    mew=1.5  # Mengatur lebar garis tepi
)

plt.title("Tren Jumlah Order Bulanan di 2018", loc="center", fontsize=24)  # Mengubah judul dan ukuran font
plt.xticks(fontsize=12, rotation=45)  # Mengubah ukuran dan rotasi label sumbu x
plt.yticks(fontsize=12)  # Mengubah ukuran label sumbu y

plt.xlabel("Bulan", fontsize=14)  # Mengubah label sumbu x dan ukuran font
plt.ylabel("Jumlah Order", fontsize=14)  # Mengubah label sumbu y dan ukuran font

plt.grid(True)  # Menambah grid
plt.tight_layout()  # Menyesuaikan layout

plt.show()
```

Grafik tersebut menunjukkan penurunan yang cukup besar pada bulan September, diikuti oleh peningkatan yang signifikan pada bulan November.

### Q 3 : Bagaimana tingkat kepuasan pelanggan?

```python
review_scores = all_data['review_score'].value_counts().sort_values(ascending=False)

most_common_score = review_scores.idxmax()

sns.set(style="darkgrid")

plt.figure(figsize=(10, 5))

sns.barplot(x=review_scores.index,
            y=review_scores.values,
            order=review_scores.index,
            palette=["#53ECEC" if score == most_common_score else "#D3D3D3" for score in review_scores.index],  # Mengubah warna menjadi #53ECEC
            edgecolor=["black" if score == most_common_score else "none" for score in review_scores.index],  # Menambahkan stroke pada warna #53ECEC
            linewidth=1.5  # Menetapkan lebar stroke
            )

plt.title("Rating Layanan dari Pelanggan", fontsize=20)  # Mengubah judul dan ukuran font
plt.xlabel("Rating", fontsize=14)  # Mengubah label sumbu x dan ukuran font
plt.ylabel("Jumlah", fontsize=14)  # Mengubah label sumbu y dan ukuran font
plt.xticks(fontsize=12)  # Mengubah ukuran label sumbu x
plt.show()
```

Grafik tersebut menunjukkan kepuasan pelanggan terhadap layanan yang disediakan, dengan mayoritas pelanggan memberikan rating 5, menunjukkan tingkat kepuasan yang tinggi.

```python
order_status_counts = all_data['order_status'].value_counts()

plt.figure(figsize=(8, 6))
order_status_counts.plot(kind='bar', color='#53ECEC', edgecolor='black', linewidth=1.5)
plt.title('Status Pemesanan', fontsize=20)
plt.xlabel('Status', fontsize=14)
plt.ylabel('Jumlah Pesanan', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

Grafik menunjukkan bahwa status order yang paling banyak adalah "delivered".

### Q 4 : Bagaimana Profil Demografi Pelanggan?

```python
# Mengelompokkan DataFrame 'all_data' berdasarkan kolom 'customer_state', kemudian menghitung jumlah unik dari 'customer_id' untuk setiap kelompok
# Mereset indeks DataFrame yang dihasilkan
# Mengubah nama kolom 'customer_id' menjadi 'customer_count'
bystate_df = all_data.groupby(by="customer_state").customer_id.nunique().reset_index()
bystate_df.rename(columns={
    "customer_id": "customer_count"
}, inplace=True)
bystate_df.head()
```

```python
all_data['order_purchase_timestamp'] = pd.to_datetime(all_data['order_purchase_timestamp'])

# Group by 'customer_id' and calculate Recency, Frequency, and Monetary
recency = (now - all_data.groupby('customer_id')['order_purchase_timestamp'].max()).dt.days
frequency = all_data.groupby('customer_id')['order_id'].count()
monetary = all_data.groupby('customer_id')['price'].sum()

# Create a new DataFrame with the calculated metrics
rfm = pd.DataFrame({
    'customer_id': recency.index,
    'Recency': recency.values,
    'Frequency': frequency.values,
    'Monetary': monetary.values
})
```

```python
# Menetapkan nama kolom baru untuk DataFrame 'rfm'
col_list = ['customer_id','Recency','Frequency','Monetary']
rfm.columns = col_list
```

```python
# Mengurutkan DataFrame 'rfm' berdasarkan kolom 'Recency' secara menaik
rfm.sort_values(by='Recency',ascending=True)
```

```python
# Mengurutkan DataFrame 'rfm' berdasarkan kolom 'Frequency' secara menurun
rfm.sort_values(by='Frequency',ascending=False)
```
